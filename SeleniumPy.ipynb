```python
import openpyxl
import datetime
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from openpyxl import utils
```

```python
# Get the current day's name
current_day = datetime.datetime.now().strftime('%A')

# Specify the path to your Excel file
file_path = 'D:\\Python\\Massudment\\Excel.xlsx'

# Load the Excel workbook
workbook = openpyxl.load_workbook(file_path)

# Get the worksheet corresponding to the current day's name
worksheet = workbook[current_day]

max_col_index = worksheet.max_column
max_row_index = worksheet.max_row
max_col_name = utils.get_column_letter(max_col_index)

print(f"Current active worksheet name is: {worksheet.title}")
print(f"Last column in this worksheet that has value is, {max_col_name} \nand last row that has value is, {max_row_index}")
```

```python
# Iterate through the columns to find the keywords and their values
keywords = []

for col_index in range(1, max_col_index + 1):
    col_name = utils.get_column_letter(col_index)
    for row_index in range(1, max_row_index + 1):
        cell_index = worksheet[f'{col_name}{row_index}']
        cell = cell_index.coordinate
        cell_value = cell_index.value
        if cell_value:
            if cell_value.lower().startswith('keyword'):
                keyword_index_column = col_name
                keyword_column = utils.get_column_letter(col_index + 1)
                keywords.append(worksheet[f'{keyword_column}{row_index}'].value)
                if cell_value.lower() == 'keyword1':
                    starting_row_index = row_index
            elif 'longest' in cell_value.lower():
                longest_keyword_column = col_name
                print(f"Cell: {cell} has the value  '{cell_value}'")

            elif 'shortest' in cell_value.lower():
                shortest_keyword_column = col_name
                print(f"Cell: {cell} has the value  '{cell_value}'")
        else:
            continue  # Skip to the next iteration
            
print(f"keywords are in column: '{keyword_column}'")
print(keywords, "\n", starting_row_index)
```

```python
# Initialize WebDriver
#driver = webdriver.Chrome(ChromeDriverManager().install())
driver = webdriver.Chrome() #--> The chromedriver.exe file is in the project folder.

for value in keywords:
    search_query = f"{value}"
    driver.get("https://www.google.com")
    search_box = driver.find_element(By.NAME, "q")
    #search_box = driver.find_element(By.XPATH, "//textarea[@id='APjFqb']")
    search_box.click()
    time.sleep(1)
    search_box.send_keys(search_query)
    search_box.send_keys(Keys.SPACE)
    time.sleep(2)
    search_suggestions = driver.find_elements(By.XPATH, "//div[@class='erkvQe']//li[@class='sbct']//div[@class='lnnVSe']")
    
    longest_suggestion = ""
    shortest_suggestion = ""
    
    for element in search_suggestions:
        keyword_suggestion = element.get_attribute("aria-label")
        print(keyword_suggestion)
        
        if not longest_suggestion or len(keyword_suggestion) > len(longest_suggestion):
            shortest_suggestion = longest_suggestion
            longest_suggestion = keyword_suggestion
        elif not shortest_suggestion or len(keyword_suggestion) < len(shortest_suggestion):
            shortest_suggestion = keyword_suggestion
    
    # Write data to Excel
    #worksheet[f'{keyword_column}{starting_row_index}'] = keywords
    #for i in range(starting_row_index, max_row):
    worksheet[f'{longest_keyword_column}{starting_row_index}'] = longest_suggestion
    worksheet[f'{shortest_keyword_column}{starting_row_index}'] = shortest_suggestion
    starting_row_index += 1

# Save the Excel workbook
workbook.save(file_path)

# Close the web driver
driver.quit()
```
